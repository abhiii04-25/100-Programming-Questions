#### 3.Write a program to read two numbers and print their sum, difference, product and quotient.

```python
num1=int(input("Enter a first number: "))
num2=int(input("Enter a second number: "))

sum=num1+num2
print("The sum of", num1, "and", num2, "is:", sum)
 
difference=num1-num2
print("The difference of", num1, "and", num2, "is:", difference)

product=num1*num2
print("The product of", num1, "and", num2, "is:", product)

quotient=num1/num2
print("The quotient of", num1, "and", num2, "is:", quotient)
```

#### 4.Write a program to read the radius of a circle and print its area and circumference.

```python
radius=float(input("Enter the radius of the circle: "))
area=3.14*radius*radius
print("The area of the circle with radius ",radius,"is:",area)

circumference=2*3.14*radius
print("The circumference of the circle with radius ",radius,"is:",circumference)
```

#### 5.Write a program to read the length and breadth of a rectangle and print its area and perimeter.

```python
lenght=float(input("Enter the length of the rectangle: "))
breath=float(input("Enter the width of the rectangle: "))
area=lenght*breath
print("Area of rectangle is: ",area)

perimeter=2(lenght+breath)
print("Perimeter of the rectangle is: ",perimeter)
```

### 6.Write a program to swap two numbers using a third variable.

```python
num1=4
num2=5
print("Befor swapping number 1 is: ",num1)
print("Befor swapping number 2 is: ",num2)
swap=num1
num1=num2
num2=swap
print("/nAfter swapping the number 1 become: ",num1)
print("/nAfter swapping the number 2 become: ",num2)
```

### 7.Write a program to swap two numbers without using a third variable.

```python
num1=4
num2=5
print("Befor swapping number 1 is: ",num1)
print("Befor swapping number 2 is: ",num2)

num1,num2=num2,num1

print("/nAfter swapping the number 1 become: ",num1)
print("/nAfter swapping the number 2 become: ",num2)
```

### 8.Write a program to read a temperature in Celsius and convert it to Fahrenheit.


```python
celsius = float(input("Enter temperature in Celsius: "))

fahrenheit = (celsius * 9/5) + 32

print(f"Temperature in Fahrenheit: {fahrenheit:.2f}°F")
```


### Program to calculate total and average marks

```python
# Program to calculate total and average marks

mark1 = float(input("Enter marks of Subject 1: "))
mark2 = float(input("Enter marks of Subject 2: "))
mark3 = float(input("Enter marks of Subject 3: "))
mark4 = float(input("Enter marks of Subject 4: "))
mark5 = float(input("Enter marks of Subject 5: "))

total = mark1 + mark2 + mark3 + mark4 + mark5
average = total / 5

print(f"\nTotal Marks: {total}")
print(f"Average Marks: {average:.2f}")
```

```python
# Program to convert seconds into hours, minutes, and seconds

seconds = int(input("Enter total seconds: "))

hours = seconds // 3600
remaining_seconds = seconds % 3600

minutes = remaining_seconds // 60
seconds = remaining_seconds % 60

print(f"Hours: {hours}")
print(f"Minutes: {minutes}")
print(f"Seconds: {seconds}")
```
